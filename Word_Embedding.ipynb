### Word Embedding Techniques using Embedding Layer in Keras

```python
### Libraries USed Tensorflow> 2.0  and keras
```

```python
##tensorflow >2.0 so that keras will be wrapped up with the tensorflow
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
### Vocabulary size
voc_size=10000
```

#### One Hot Representation

```python
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr) # for index representation
```

### Word Embedding Represntation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8 # based on your sentence length
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length) # pre add 0's in front to make sentences complete for 8 word sentence
print(embedded_docs) # for formatting index matrix
```

```python
dim=10 # 10 feature representation
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
